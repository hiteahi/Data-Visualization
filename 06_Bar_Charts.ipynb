# Bar Chart in Matplotlib
Bar plots are significant because they provide a clear and intuitive way to visualize categorical data. They allow viewers to quickly grasp differences in size or quantity among categories, making them ideal for presenting survey results, sales data, or any discrete variable comparisons.

```python
import pandas as pd, numpy as np
import matplotlib.pyplot as plt

# simple dataset
month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
apple = [15000, 18000, 12000, 22000, 21000, 25000, 28000, 24000, 31000, 35000, 48000, 52000]
banana = [9000, 11800, 6200, 15500, 14600, 18400, 19000, 17500, 24000, 27800, 39500, 43000]
mango = [12000, 15000, 11000, 19000, 18500, 21000, 24000, 20000, 28000, 31000, 45000, 50000]
```

## Basic Bar Plot

```python
plt.bar(month, apple)
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.title("Apple's Sales")
# plt.show()
```

### Bar Color

The `bar()` and `barh()` take the keyword argument color to set the color of the bars

We can set color by color name and color hex

```python
plt.bar(month, apple, color="#139418") # color by hex color code
plt.ylabel("Revenue")
plt.title("Apple's Sales")
plt.grid(True, alpha=0.2)
```

### Bar Width

```python
plt.bar(month, apple, color="lightgreen", width=0.5) # width is 0.5 & color by name
plt.ylabel("Revenue")
plt.title("Aplle's Sales")
plt.grid(True, alpha=0.2)
```

The value of width ranges from 0 to 1 (default 0.8); it can go above 1, but bars will start to overlap.

## Horizontal Bars

Horizontal bar charts are better when category names are long.

If you want the bars to be displayed horizontally instead of vertically, use the `barh()` function

```python
plt.barh(month, banana, color="#DBD400")
plt.xlabel("Revenue")
plt.title("Banana's Sales")
plt.grid(True, alpha=0.2)
```

### Bar Height

Use height instead of width in `barh()`

```python
plt.barh(month, mango, color="#DBAF00", height=0.5)
plt.xlabel("Revenue")
plt.title("Mango's Sales")
```

## Multiple bar plots
Setting Bar Width & Side-by-Side Bar Charts

```python
x = np.arange(len(month)) # index positions
width = 0.25 

plt.bar(x - width, apple, width=width, label="Apple", color="green")
plt.bar(x, banana, width=width, label="Banana", color="yellow")
plt.bar(x + width, mango, width=width, label="Mango", color="red")

plt.ylabel("Sales Volume")
plt.title("Fruits Sales")
plt.legend()
plt.tight_layout()
```

##  Use xticks() & ysticks()

By default, `plt.bar()` uses numeric x-values (0, 1, 2, ...).

We use `plt.xticks()` and `plt.yticks()` to set the correct category labels like months or names

```python
x = np.arange(len(month)) 
width = 0.25 

plt.bar(x - width, apple, width=width, label="Apple", color="green")
plt.bar(x, banana, width=width, label="Banana", color="yellow")
plt.bar(x + width, mango, width=width, label="Mango", color="red")

plt.ylabel("Sales Volume")
plt.title("Fruits Sales")
plt.xticks(x, month) # Show actual year instead of 0,1,2,...
plt.legend()
plt.tight_layout()
```

```python
y = np.arange(len(month)) 
h = 0.25 # height of the horizontal bar

# Pass 'height' as a keyword argument
plt.barh(y - h, apple, height=h, label="Apple", color="green")
plt.barh(y, banana, height=h, label="Banana", color="yellow")
plt.barh(y + h, mango, height=h, label="Mango", color="red")

plt.xlabel("Sales Volume") 
plt.title("Fruits Sales")
plt.yticks(y, month) 
plt.legend()
plt.tight_layout()
```

## Adding Value Labels with plt.text()
We can use `plt.text()` to display values on top of bars, making your chart easier to read. It's especially
useful in presentations and reports.

```python
# simple dataset
fruits = ["Apple", "Banana", "Mango"]
sales = [331000, 246300, 294500]

plt.bar(fruits, sales, color="skyblue")

# Add labels on top of bars
for i in range(len(sales)):
    plt.text(i, sales[i] + 50, str(sales[i]), ha='center') 


plt.ylabel("Sales")
plt.title("Fruits Sales")
plt.tight_layout()
plt.show()
```

**Tip**: Use ha='center' to center the text and add a small offset (+50 here) to avoid overlap with the bar

## Summary


| Feature | Use |
| :--- | :--- |
| `plt.bar()` | Vertical bars for categorical comparison |
| `plt.barh()` | Horizontal bars (great for long labels) |
| `width=` | Control thickness/spacing of bars |
| `np.arange()` | Helps to align multiple bars side by side |
| `plt.xticks()` | Replaces index numbers with real labels |
| `plt.tight_layout()` | Prevents labels from overlapping |
