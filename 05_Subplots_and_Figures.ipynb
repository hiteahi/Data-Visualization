# Subplot in Matplotlib

Subplots are commonly used in dashboards
to compare metrics side by side.

```python
import pandas as pd, numpy as np
import matplotlib.pyplot as plt

# simple dataset
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Revenue': [15000, 18000, 12000, 22000, 21000, 25000, 28000, 24000, 31000, 35000, 48000, 52000],
    'Customers': [120, 150, 110, 190, 185, 210, 240, 200, 280, 310, 450, 500],
    'Ad_Spend': [1000, 1200, 800, 1500, 1400, 1600, 1800, 1500, 2000, 2200, 3500, 4000]
}

df = pd.DataFrame(data)

# Assuming fixed monthly costs are 5,000
fixed_costs = 5000

# Calculate Profit: Revenue - (Ad_Spend + Fixed_Costs)
df['Profit'] = df['Revenue'] - (df['Ad_Spend'] + fixed_costs)
```

## Multiple Plots with `subplot()` Function

Subplots are useful when comparing multiple trends,
such as sales across different regions.

With the `subplot()` function we can draw multiple plots in one figure

- The `subplot()` function takes three arguments that describes the layout of the figure.
- The layout is organized in rows and columns, which are represented by the first and second argument.
- The third argument represents the index of the current plot.

```python
# Plot 1
plt.subplot(2,1,1) # 2 for row, 1 for columns and this plot is first plot.
plt.plot(df["Month"], df["Revenue"])
plt.ylabel("Revenue")

# Plot 2
plt.subplot(2,1,2) # 2 for row, 1 for columns and this plot is second plot.
plt.plot(df["Month"], df["Profit"])
plt.ylabel("Profit")
```

```python
# Plot 1
plt.subplot(1,2,1) # 1 for row, 2 for columns and this plot is first plot.
plt.plot(df["Revenue"], df["Month"])
plt.ylabel("Revenue")

# Plot 2
plt.subplot(1,2,2) # 1 for row, 2 for columns and this plot is second plot.
plt.plot(df["Revenue"], df["Month"])
plt.ylabel("Profit")
```

## Super Title
We can add a title to the entire figure with the `suptitle()` function

```python
# Plot 1
plt.subplot(2,1,1) 
plt.plot(df["Month"], df["Revenue"])
plt.ylabel("Revenue")

# Plot 2
plt.subplot(2,1,2) 
plt.plot(df["Month"], df["Profit"])
plt.ylabel("Profit")

plt.suptitle("Revenue & Profit")
```

```python
# Plot 1
plt.subplot(2,2,1) 
plt.plot(df["Revenue"])
plt.ylabel("Revenue")

# Plot 2
plt.subplot(2,2,2) 
plt.plot(df["Profit"])
plt.ylabel("Profit")

# Plot 3
plt.subplot(2,2,3) 
plt.plot(df["Customers"])
plt.ylabel("Customers")

# Plot 4
plt.subplot(2,2,4) 
plt.plot(df["Ad_Spend"])
plt.ylabel("Ad Spend")

# tight_layout() prevents overlapping labels and improves spacing.
plt.tight_layout()
plt.suptitle("Monthly Data")
```

## Figure (fig)
The top-level container for all plot elements (titles, legends, and subplots).

```python
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(df["Revenue"], df["Month"])
axs[0, 0].set_title("Revenue")
axs[0, 1].plot(df["Profit"], df["Month"])
axs[0, 1].set_title("Profit")
axs[1, 0].plot(df["Customers"], df["Month"])
axs[1, 0].set_title("Customers")
axs[1, 1].plot(df["Ad_Spend"], df["Month"])
axs[1, 1].set_title("Ad Spend")
plt.tight_layout() # Adjusts spacing to prevent overlap
fig.suptitle("Monthly Data")
```

## Summary

- A figure can contain multiple axes
- subplots() creates structured layouts
- tight_layout() improves spacing
- Subplots are useful for comparison
