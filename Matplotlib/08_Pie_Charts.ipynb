# Pie Charts in Matplotlib

Pie charts are used to show part-to-whole relationships. They're visually appealing but best used with fewer
categories, as too many slices can get cluttered and hard to interpret.

## Pie charts are not suitable when:
- There are too many categories
- Differences between values are small
- Precise comparison is required

```python
import pandas as pd, numpy as np
import matplotlib.pyplot as plt

# simple dataset
fruits = ["Apple", "Banana", "Mango", "Grapes"]
sales = [405000, 206300, 294500, 88500]
```

```python
plt.pie(sales, labels=fruits)
plt.title("Fruits Sales")
plt.show()
```

## Custom Colors

```python
colors = ['#e03a3e','#FFE135','#F4BB44','#6F2DA8']
plt.pie(sales, labels=fruits, colors=colors)
plt.title("Fruits Sales")
```

## Donut Chart

```python
colors = ['#e03a3e','#FFE135','#F4BB44','#6F2DA8']
plt.pie(sales, labels=fruits, colors=colors)
plt.gca().add_artist(plt.Circle((0,0),0.3,fc='white'))
plt.title("Fruits Sales")
```

## Add Edges and Style Slices with wedgeprops

We can customize the look of the slices using `wedgeprops`.

```python
plt.pie(
 sales,
 labels=fruits,
 colors=colors,
 wedgeprops={'edgecolor': 'white', 'linewidth': 2, 'linestyle': ':'} 
)
plt.title("Fruits Sales")
```

## Highlight a Slice Using explode

Use `explode` to pull out one or more slices.

```python
explode = [0.1, 0, 0, 0] # Only highlight Apple slice
plt.pie(sales, 
        labels=fruits, 
        explode=explode, 
        colors=colors
)
plt.title("Fruits Sales")
```

## Add Shadows for a 3D Feel

```python
plt.pie(
 sales,
 labels=fruits,
 explode=explode,
 colors=colors,
 shadow=True # adds a 3D-like shadow
)
plt.title("Fruits Sales")
```

## Start Angle

Use `startangle` to rotate the pie for better alignment.

```python
plt.pie(
 sales,
 labels=fruits,
 explode=explode,
 colors=colors,
 shadow=True,
 startangle=300     # rotate with 300 
)
plt.title("Fruits Sales")
```

## Show Percentages with `autopct`

```python
plt.pie(
 sales,
 labels=fruits,
 explode=explode,
 colors=colors,
 shadow=True,
 startangle=300,
 autopct='%1.1f%%' # adding parcentages
)
plt.title("Fruits Sales")
```

## Hatch slices
Pass a list of hatch patterns to hatch to set the pattern of each slice.

```python
hatch_patterns = ['*', 'o', '+', '.']
plt.pie(
 sales,
 labels=fruits,
 explode=explode,
 colors=colors,
 shadow=True,
 startangle=300,
 hatch=hatch_patterns
)
plt.title("Fruits Sales")
```

## Crowded Pie Chart – Avoid Too Many Categories

```python
# simple dataset
fruits = ["Apple", "Banana", "Mango", "Grapes", "Guava", "Papaya", "Orange", "Pineapple"]
sales = [355, 206, 294, 158, 195, 258, 210, 310]

plt.pie(sales, labels=fruits, autopct='%1.1f%%')
plt.title("Fruits Sales")
```

## Why avoid it?

- Hard to compare slice sizes visually
- Cluttered and confusing
- No clear insight

**Better alternatives**: bar charts or horizontal bar charts

## Summary


| Parameter | Description |
| :--- | :--- |
| **labels** | A list of strings providing a name for each slice. |
| **colors** | A list of colour names or Hex codes to style the slices. |
| **explode** | An array of values (e.g., 0.1) that offsets a slice from the centre. |
| **shadow** | A boolean (True/False) that adds a 3D-like drop shadow. |
| **startangle** | The degree at which the first slice starts (e.g., 90 for the top). |
| **autopct** | A string/function used to label the slices with their numeric value. |
| **wedgeprops** | A dictionary of properties (like linewidth or edgecolor) for slice borders. |
