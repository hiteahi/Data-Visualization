# Scatter in Matplotlib

Scatter plots are used to show the relationship between two variables. 

```python
import pandas as pd, numpy as np
import matplotlib.pyplot as plt

# simple dataset
month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
apple = [15000, 18000, 12000, 22000, 21000, 25000, 28000, 24000, 31000, 35000, 48000, 52000]
banana = [9000, 11800, 16200, 18500, 23600, 18400, 19000, 17500, 29000, 27800, 39500, 43000]
mango = [12000, 15000, 15000, 23000, 25500, 21000, 24000, 20000, 28000, 31000, 45000, 50000]
```

## Basic Scatter Plot

```python
plt.scatter(month, apple)
plt.title("Apple's Sales")
plt.ylabel('Apple')
plt.grid(True, alpha=0.5)
plt.show()
```

This shows a simple scatter plot. You can notice with time sales increase and decrease but overall sales increase.

## Adding Color and Size

```python
# Size of points based on score (higher sales -> bigger point)
sizes = [score * 0.01  for score in apple]
colors = ['red' if score < 30000 else 'green' for score in apple]

plt.scatter(month, apple, s=sizes, c=colors, alpha=0.7)
plt.title('Colored & Sized Scatter Plot')
plt.ylabel('Apple')
plt.grid(True, alpha=0.5)
```

Points are colored based on performance and scaled in size by score. Red means low score, green
means good.

## Using a Colormap

```python
sizes = [score * 0.01  for score in apple]

plt.scatter(month, apple, s=sizes, c=apple, edgecolors='w', alpha=0.8, cmap='viridis')
plt.colorbar(label='Sales Volume')
plt.title('Sales Visualization with Colormap')
plt.ylabel('Apple')
plt.grid(True, alpha=0.5)
```

`cmap` adds gradient coloring based on the score. `colorbar` helps understand what the colors
represent

## Adding Annotations

```python
plt.scatter(month, apple, s=[s*0.01 for s in apple], c=apple, cmap='cool', alpha=0.75)

# Annotating ALL points using a loop
for i, txt in enumerate(apple):
    plt.annotate(txt, (month[i], apple[i]), 
                 textcoords="offset points", # Position text relative to point
                 xytext=(0,10),              # Distance from point (x,y)
                 ha='center',                # Horizontal alignment
                 fontsize=8)
    
plt.title('Sales with Annotations')
plt.ylabel('Apple')
plt.grid(True, alpha=0.25)
```

Annotations help identify individual points, which is useful in small datasets

## Compare Plots
Multiple Groups in One Plot

```python
plt.scatter(month, apple, label='Apple', color='#af3e4d')
plt.scatter(month, banana, label='Banana', color='#DBD400')
plt.scatter(month, mango, label='Mango', color='#DBAF00')
plt.title('Fruits Sales')
plt.legend()
plt.grid(True, alpha=0.5)
```

When comparing two or multiple datasets, use different colors and a legend for clarity

## Summary

| Feature | Use |
| :--- | :--- |
| `plt.scatter(x, y)` | Creates a scatter plot with individual point controln |
| `s=` | Controls Size of points (can be a single value or an array) |
| `c=` | Controls Color (accepts color names or numeric data) |
| `cmap=` | Applies a Colormap (e.g., 'viridis', 'magma', 'Reds') |
| `alpha=` | Sets Transparency (0 to 1) to handle overlapping points |
| `marker=` | Changes the Shape (e.g., 'o', 's', '^', '*') |
| `edgecolors=` | Adds a border color to markers for better visibility |
| `plt.colorbar()` | Adds a visual scale legend for the colormap |
| `plt.annotate()` | Adds text labels or highlights to specific data points |
| `plt.plot(x, y, 'o')` | A faster alternative for very large datasets without size/color variation |
