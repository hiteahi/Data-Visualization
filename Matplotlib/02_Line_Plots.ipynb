# Plot in Matplotlib

```python
import pandas as pd, numpy as np
import matplotlib.pyplot as plt
```

```python
# simple dataset
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Revenue': [15000, 18000, 12000, 22000, 21000, 25000, 28000, 24000, 31000, 35000, 48000, 52000],
    'Customers': [120, 150, 110, 190, 185, 210, 240, 200, 280, 310, 450, 500],
    'Ad_Spend': [1000, 1200, 800, 1500, 1400, 1600, 1800, 1500, 2000, 2200, 3500, 4000]
}

df = pd.DataFrame(data)

# Assuming fixed monthly costs are 5,000
fixed_costs = 5000

# Calculate Profit: Revenue - (Ad_Spend + Fixed_Costs)
df['Profit'] = df['Revenue'] - (df['Ad_Spend'] + fixed_costs)
```

## Plotting x and y points

The `plot()` function is used to draw points (markers) in a diagram.

By default, the `plot()` function draws a line from point to point.

The function takes parameters for specifying points in the diagram.

**Parameter 1** is an array containing the points on the **x-axis**.

**Parameter 2** is an array containing the points on the **y-axis**.

```python
plt.plot(df["Month"], df["Revenue"])
plt.show() # Not needed in jupyter notebook 
```

**Note**: By default, matplotlib draws a line plot when using plot()

## Plotting Without Line
To plot only the markers, we can use shortcut string notation parameter `'o'`, which means **'rings'**.

```python
plt.plot(df["Month"], df["Revenue"], "o")
```

## Default X-Points
If we do not specify the points on the x-axis, they will get the default values 0, 1, 2, 3 etc., depending on the length of the y-points.

```python
plt.plot(df["Revenue"])
```

## Line Styles

We can use the keyword argument `linestyle`, or shorter `ls`, to change the style of the plotted line

```python
plt.plot(df["Month"], df["Revenue"], linestyle=':') # "linestyle=':'" means Dotted Line
```

choose any of these styles:

| Style | Or |
| :--- | :--- |
| `'solid'` (default) | `'-'` |
| `'dotted'` | `':'` |
| `'dashed'` | `'--'` |
| `'dashdot'` | `'-.'` |
| `'None'` | `''` or `' '` |

## Line Color
We can use the keyword argument `color` or the shorter `c` to set the color of the line

```python
plt.plot(df["Month"], df["Revenue"], color='#990099') # We can also use Hexadecimal color
```

## Line Width
We can use the keyword argument `linewidth` or the shorter `lw` to change the width of the line.

```python
plt.plot(df["Month"], df["Revenue"], linewidth=8) # "linewidth=8" means Line Width is 8pt
```

## Multiple Lines in One Plot
We can plot as many lines as you like by simply adding more `plt.plot()` functions

```python
plt.plot(df["Month"], df["Revenue"], label="Revenue")
plt.plot(df["Month"], df["Profit"], label="Profit") # Adding Profit Line
```

## Adding Legend
Legends help distinguish multiple lines.

```python
plt.plot(df["Month"], df["Revenue"], label="Revenue")
plt.plot(df["Month"], df["Profit"], label="Profit")
plt.legend() # Adding Legend
```

## Summary

- Line plots show trends over continuous data
- Multiple lines allow comparison
- Proper labeling and legends are essential
