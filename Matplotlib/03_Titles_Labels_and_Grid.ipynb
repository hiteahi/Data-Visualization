# Title, Labels and Grid in Matplotlib

```python
import pandas as pd, numpy as np
import matplotlib.pyplot as plt
```

```python
# simple dataset
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Revenue': [15000, 18000, 12000, 22000, 21000, 25000, 28000, 24000, 31000, 35000, 48000, 52000],
    'Customers': [120, 150, 110, 190, 185, 210, 240, 200, 280, 310, 450, 500],
    'Ad_Spend': [1000, 1200, 800, 1500, 1400, 1600, 1800, 1500, 2000, 2200, 3500, 4000]
}

df = pd.DataFrame(data)
```

## Adding Title in Plot
We can use the `title()` function to set a title for the plot.

```python
plt.plot(df["Month"], df["Revenue"])
plt.title("Monthy Revenue") # Plotting with Title
```

## Adding X and Y Labels in Plot
We can use the `xlabel()` and `ylabel()` functions to set a label for the x- and y-axis.

```python
plt.plot(df["Month"], df["Revenue"])
plt.xlabel("Month") # X Label
plt.ylabel("Revenue") # Y Label
plt.title("Sales Over Time", fontsize=18) # "fontsize=18" means Title font size is 18
```

**Note**: A chart without labels and title can confuse viewers. Proper labeling makes the visualization self-explanatory.

## Font Properties
We can use the `fontdict` parameter in `xlabel()`, `ylabel()`, and `title()` to set font properties for the title and labels.

```python
font1 = {'family':'serif','color':'green','size':20}
font2 = {'family':'serif','color':'blue','size':15}

plt.plot(df["Month"], df["Revenue"])
plt.xlabel("Month", fontdict = font2) 
plt.ylabel("Revenue",  fontdict = font2)
plt.title("Monthy Sales", fontdict = font1)
```

## Position the Title

We can use the `loc` parameter in `title()` to position the title.

Legal values are: `'left'`, `'right'`, and `'center'`. Default value is `'center'`.

```python
plt.plot(df["Month"], df["Revenue"])
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.title("Monthy Sales", loc="left", fontdict = font1) # Title in Left side
```

## Adding Grid Lines

Grid improves readability and value estimation.

```python
plt.plot(df["Month"], df["Revenue"]) 
plt.xlabel("Month") 
plt.ylabel("Revenue")
plt.title("Monthy Sales", fontsize=20)
plt.grid() # Adding Grid Lines
```

## Specify Which Grid Lines
we can use the axis parameter in the grid() function to specify which grid lines to display.

### Grid lines for the x-axis

```python
plt.plot(df["Month"], df["Revenue"])
plt.xlabel("Month") 
plt.ylabel("Revenue")
plt.title("Monthy Revenue")
plt.grid(axis='x') # only x-axis grid
```

### Grid lines for the y-axis

```python
plt.plot(df["Month"], df["Revenue"])
plt.xlabel("Month") 
plt.ylabel("Revenue")
plt.title("Monthy Sales")
plt.grid(axis='y') # only y-axis grid
```

## Line Properties for Grid

```python
plt.plot(df["Month"], df["Revenue"], linewidth=10)
plt.xlabel("Month") 
plt.ylabel("Revenue")
plt.title("Monthy Sales")
plt.grid(color = 'purple', linestyle = ':', linewidth = 2)
```

```python
plt.plot(df["Month"], df["Revenue"], linewidth=10)
plt.xlabel("Month") 
plt.ylabel("Revenue")
plt.title("Monthy Sales")
plt.grid(True, linestyle='--', alpha=0.5)
```

## Summary

- Titles explain what the chart represents
- Axis labels clarify units and meaning
- Grids improve readability
- Well-labeled charts communicate effectively
