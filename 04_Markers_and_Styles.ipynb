# Markers and Styles in Matplotlib

```python
import pandas as pd, numpy as np
import matplotlib.pyplot as plt

# simple dataset
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Revenue': [15000, 18000, 12000, 22000, 21000, 25000, 28000, 24000, 31000, 35000, 48000, 52000],
    'Customers': [120, 150, 110, 190, 185, 210, 240, 200, 280, 310, 450, 500],
    'Ad_Spend': [1000, 1200, 800, 1500, 1400, 1600, 1800, 1500, 2000, 2200, 3500, 4000]
}

df = pd.DataFrame(data)

# Assuming fixed monthly costs are 5,000
fixed_costs = 5000

# Calculate Profit: Revenue - (Ad_Spend + Fixed_Costs)
df['Profit'] = df['Revenue'] - (df['Ad_Spend'] + fixed_costs)
```

## Markers

Markers are useful when the number of data points is small
or when highlighting individual observations is important.

We can use the keyword argument `marker` to emphasize each point with a specified marker

```python
plt.plot(df["Month"], df["Revenue"], marker="d") # d for diamond
plt.xlabel("Month") 
plt.ylabel("Revenue")
plt.title("Monthy Sales")
```

## Marker Size
We can use the keyword argument `markersize` or the shorter version, `ms` to set the size of the markers

```python
plt.plot(df["Month"], df["Revenue"], marker="d", ms=15) # "ms=15" means Marker Size is 15
plt.xlabel("Month") 
plt.ylabel("Revenue")
plt.title("Monthy Sales")
```

## Marker Edge Color
We can use the keyword argument `markeredgecolor` or the shorter `mec` to set the color of the edge of the markers

```python
plt.plot(df["Month"], df["Revenue"], marker="d", ms=15, mec='r') # "mec='r'" means Marker Edge Color is Red
plt.xlabel("Month") 
plt.ylabel("Revenue")
plt.title("Monthy Sales")
```

## Marker FACE Color
We can use the keyword argument `markerfacecolor` or the shorter `mfc` to set the color inside the edge of the markers

```python
plt.plot(df["Month"], df["Revenue"], marker="d", ms=15, mec="r", mfc='g') # "mfc='g'" means Marker FACE Color is Green
plt.xlabel("Month") 
plt.ylabel("Revenue")
plt.title("Monthy Sales")
```

```python
# Use both the mec and mfc arguments to color the entire marker
plt.plot(df["Month"], df["Revenue"], marker="d", ms=15, mec="hotpink", mfc='hotpink')
plt.xlabel("Month") 
plt.ylabel("Revenue")
plt.title("Monthy Sales")
```

## Marker + Linestyle

```python
plt.plot(df["Month"], df["Revenue"], marker='o', linestyle='--')
```

**Note**: Overusing colors and marker styles can make charts cluttered and reduce readability.

## Format Strings `fmt`
We can also use the shortcut string notation parameter to specify the marker.
This parameter is also called fmt.

```python
plt.plot(df["Month"], df["Revenue"], "ro--", label="Revenue") # red circles with dashed lines
plt.plot(df["Month"], df["Profit"], "g^:", label="Profit") # green triangles dotted
plt.xlabel("Month") 
plt.ylabel("Revenue")
plt.title("Monthy Sales")
plt.legend()
```

## Using Styles in Matplotlib

```python
# check available styles in matplotlib
plt.style.available
```

## Using ggplot Style

```python
plt.style.use("ggplot")
plt.plot(df["Month"], df["Revenue"], label="Revenue")
plt.plot(df["Month"], df["Profit"], label="Profit") 
plt.xlabel("Month") 
plt.ylabel("Revenue")
plt.title("Monthy Sales")
plt.legend()
```

## Temporarily apply a style
Use a context manager to apply a style only within a specific block of code

```python
with plt.style.context('dark_background'):
    plt.plot(df["Month"], df["Revenue"], label="Revenue")
    plt.plot(df["Month"], df["Profit"], label="Profit") 
    plt.xlabel("Month") 
    plt.ylabel("Revenue")
    plt.title("Monthy Sales")
    plt.legend()
```
