# Stack Plots in Matplotlib

A Stack Plot is a type of area chart that helps visualize multiple quantities over time, stacked on top of each
other. It's especially useful to see how individual parts contribute to a whole over time.


| Feature | Best Used For |
| :--- | :--- |
| **Data Type** | Time-series (Days, Months, Years) |
| **Goal** | Showing how the "Total" is made up of "Parts" |
| **Avoid When...** | You have more than 5 categories (it gets too messy) |
| **Avoid When...** | The values fluctuate wildly (makes the stacks look like "waves") |


```python
import matplotlib.pyplot as plt
import pandas as pd, numpy as np

# Days of the week
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']

# Hours spent on each activity
working = [8, 8, 9, 8, 7, 2, 0]
sleeping = [7, 7, 6, 7, 8, 9, 9]
gaming = [2, 3, 2.5, 2, 4, 5, 6]
relaxing = [7, 6, 6.5, 7, 5, 8, 11]

activities = [working, sleeping, relaxing, gaming]
labels = ['Work', 'Sleep', 'Relax', 'Games']
colors = ['skyblue', 'lightgreen', 'gold', 'lightcoral']
```

## Basic Stack Plot

```python
plt.figure(figsize=(10, 6))
plt.stackplot(days, activities, labels=labels, colors=colors, alpha=0.7)
plt.title('Weekly Time Distribution', fontsize=15)
plt.ylabel('Hours in a Day')
plt.legend(loc='upper left')
plt.show()
```

## Edge Colors
a thin line between the stacks to make them distinct.

```python
plt.figure(figsize=(10, 6))
plt.stackplot(days, activities, labels=labels, colors=colors, edgecolor='white', linewidth=0.5, alpha=0.75)
plt.title('Weekly Time Distribution', fontsize=15)
plt.ylabel('Hours in a Day')
plt.legend(loc='upper left')
```

## Baseline
Ensure the Y-axis starts at 0 and goes to 24

```python
plt.figure(figsize=(10, 6))
plt.stackplot(days, activities, labels=labels, colors=colors, edgecolor='white', linewidth=0.5, alpha=0.75)
plt.ylim(0, 24)
plt.title('Weekly Time Distribution', fontsize=15)
plt.ylabel('Hours in a Day')
plt.legend(loc='upper left')
```

## Key Parameters in `stackplot()`

| Parameter | Description |
| :--- | :--- |
| **x** | The x-axis data (like days). |
| **args** | Multiple y-values (like studying, playing, etc.). |
| **labels** | Labels for the legend. |
| **colors** | List of colors for each stack. |
| **alpha** | Transparency level (0 to 1). |
| **loc (in legend)** | Position of the legend (e.g., 'upper left', 'best'). |
